1. User Login Check

```python
stored_username="admin"
stored_password="1234"

username = input("Enter your username:")
password = input("Enter your password:")

if username == stored_username and password == stored_password:
  print("Login Successful")
else:
  print("Invalid Credentials")
```

2. Pass / Fail Analyzer

```python
marks = [45,78,90,33,60]

pass_count = 0
fail_count = 0

for mark in marks:
  if mark >=50:
    pass_count += 1
  else:
    fail_count += 1
print("FINAL RESULT :-")
print("Total number of Students:", len(marks))
print("Number of passed Students:",pass_count)
print("Number of failed Students:",fail_count)
```

3. Simple Data Cleaner


```python
names =["Alice","bob","CHARLIE"]

cleaned_names = []

for name in names:
  cleaned_name = name.strip().lower()
  cleaned_names.append(cleaned_name)
print("Previous Names List",names)
print("Cleaned Names List",cleaned_names)
```

4. Message Length Analyzer



```python
messages = ["Hi", "Welcome to the platform", "OK"]


for message in messages:
    length = len(message)
    print(f"Message: '{message}'")
    print(f"Length: {length}")

    if length > 10:
        print("Status: FLAGGED!! Longer than 10 characters")
    else:
        print("Status: Within limit")

    print()
```

5. Error Message Detector


```python
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

error_count = 0

for log in logs:
    if log == "ERROR":
        error_count += 1

print("Total ERROR entries:", error_count)
```
