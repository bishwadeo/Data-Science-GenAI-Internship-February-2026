1. Unique Words in a Sentence


```python
sentence = input("Enter sentence: ")
words = sentence.split()
unique_words = set(words)

print("Unique words:", unique_words)
print("Count of unique words:", len(unique_words))
```

2. Highest Salary from Employee Data



```python
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

highest_paid = max(employees, key=employees.get)

print("Employee with highest salary:", highest_paid)
print("Highest salary:", employees[highest_paid])
```

3. Find Maximum and Minimum Values


```python
numbers = [45, 22, 89, 10, 66]

maximum = numbers[0]
minimum = numbers[0]

for num in numbers:
    if num > maximum:
        maximum = num
    if num < minimum:
        minimum = num
print("list:",numbers)
print("Maximum value:", maximum)
print("Minimum value:", minimum)
```

4. Count Products Above a Price Threshold




```python
prices = [450, 1200, 899, 1500, 300]

count = 0

for price in prices:
    if price > 1000:
        count += 1

print("Number of products with price greater than 1000:", count)
```

5. Calculate Attendance Percentage



```python
attendance = ["P", "P", "A", "P", "P"]

present_count = 0

for day in attendance:
    if day == "P":
        present_count += 1

total_days = len(attendance)

percentage = (present_count / total_days) * 100

print("Attendance Percentage:", percentage, "%")
```

6. Remove Duplicate Phone Numbers


```python
phone_numbers = [9876543210, 9123456789, 9876543210]

unique_numbers = set(phone_numbers)

print("Unique phone numbers:",unique_numbers)
```

7. Count Character Frequency


```python
string = "pythonp"

char_count = {}

for char in string:
    if char in char_count:
        char_count[char] += 1
    else:
        char_count[char] = 1

print(char_count)
```

8. Convert List to Tuple


```python
numbers = [10, 20, 30]

numbers_tuple = tuple(numbers)

print("Tuple:", numbers_tuple)
```

09. Check if a Key Exists in a Dictionary


```python
employees = {
    "Dev": 25000,
    "Aman": 28000,
    "Priya": 30000
}
key_to_check = "Dev"

if key_to_check in employees:
    print("Employee exists")
```

10. Calculate Average Marks



```python
marks_input = input("Enter marks: ")

marks = [int(mark) for mark in marks_input.split()]

total_marks = sum(marks)
num_subjects = len(marks)
average = total_marks / num_subjects

print("Average marks:", average)
```
