1. Mobile Recharge Validation System


```python
def validate_recharge():
    valid_plans = [199, 299, 399, 599]

    while True:
        try:
            amount = int(input("Enter recharge amount: ₹"))
        except ValueError:
            print("Invalid input! Please enter a number.")
            continue

        if amount < 50:
            print("Recharge amount must be at least ₹50. Try again.")
        elif amount not in valid_plans:
            print(f"Invalid plan! Valid plans are: {valid_plans}. Try again.")
        else:
            print(f"Recharge of ₹{amount} successful!")
            return amount  # Return valid amount

# Example usage
recharge_amount = validate_recharge()
```

2. Inventory Reorder Alert System


```python
def check_stock(products):
    for product, quantity in products.items():
        if quantity < 15:
            print(f"{product}: Reorder Alert")
        else:
            print(f"{product}: Stock OK")

# Example usage
inventory = {
    "Apples": 20,
    "Bananas": 10,
    "Oranges": 5,
    "Mangoes": 25
}

check_stock(inventory)
```

3. Student Result Processing System


```python
def check_result(marks):
    total = 0
    for mark in marks:
        total += mark
    average = total / len(marks)

    if average >= 50:
        print("Pass")
    else:
        print("Fail")

    return average

# Example usage
student_marks = [60, 70, 55, 40, 50]
avg = check_result(student_marks)
print("Average marks:", avg)
```

4. Cab Fare Estimator with Retry Option


```python
def calculate_fare(distance_km, peak_hour=False):
    base_fare = 50
    per_km_rate = 12
    fare = base_fare + (per_km_rate * distance_km)

    if peak_hour:
        fare *= 1.25  # 25% extra

    return round(fare, 2)

# Retry loop
while True:
    try:
        distance = float(input("Enter distance in km: "))
        peak = input("Is it peak hour? (yes/no): ").strip().lower()
        is_peak = peak == "yes"

        if distance <= 0:
            print("Distance must be positive. Try again.")
            continue

        total_fare = calculate_fare(distance, is_peak)
        print(f"Total fare: ₹{total_fare}")
        break  # Exit loop after valid input
    except ValueError:
        print("Invalid input. Please enter a number for distance.")
```

5. Employee Attendance Eligibility

```python
def check_attendance(attendance_list):
    total_days = len(attendance_list)
    present_days = 0

    for status in attendance_list:
        if status.upper() == "P":
            present_days += 1

    attendance_percent = (present_days / total_days) * 100

    if attendance_percent >= 75:
        return "Eligible"
    else:
        return "Not Eligible"

# Example usage
attendance = ["P", "A", "P", "P", "A", "P", "P", "P"]
result = check_attendance(attendance)
print("Attendance result:", result)
```

6. Password Strength Checker


```python
def check_password(password):
    special_chars = "@#$"

    if len(password) < 8:
        return "Weak: Password too short"

    has_digit = False
    has_special = False

    for char in password:
        if char.isdigit():
            has_digit = True
        elif char in special_chars:
            has_special = True

    if has_digit and has_special:
        return "Strong"
    elif not has_digit and not has_special:
        return "Weak: Missing digit and special character"
    elif not has_digit:
        return "Weak: Missing digit"
    else:
        return "Weak: Missing special character"

# Example usage
passwords = ["mypassword", "mypassword1", "mypassword@", "mypassword1@", "Pass1@"]
for pwd in passwords:
    print(f"{pwd}: {check_password(pwd)}")
```
